### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingresos_total_t1 = tienda['Precio'].sum()
ingresos_total_t2 = tienda2['Precio'].sum()
ingresos_total_t3 = tienda3['Precio'].sum()
ingresos_total_t4 = tienda4['Precio'].sum()
```

```python
print("El ingreso total para la tienda 1 es: ", ingresos_total_t1)
print("El ingreso total para la tienda 2 es: ", ingresos_total_t2)
print("El ingreso total para la tienda 3 es: ", ingresos_total_t3)
print("El ingreso total para la tienda 4 es: ", ingresos_total_t4)
```

# 2. Ventas por categoría

```python
ventas_categoria_t1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
ventas_categoria_t2 = tienda.groupby('Categoría del Producto')['Precio'].sum()
ventas_categoria_t3 = tienda.groupby('Categoría del Producto')['Precio'].sum()
ventas_categoria_t4 = tienda.groupby('Categoría del Producto')['Precio'].sum()
```

```python
print("Las ventas por categoría para la tienda 1 son: \n", ventas_categoria_t1)
print("Las ventas por categoría para la tienda 2 son: \n", ventas_categoria_t2)
print("Las ventas por categoría para la tienda 3 son: \n", ventas_categoria_t3)
print("Las ventas por categoría para la tienda 4 son: \n", ventas_categoria_t4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_prom_t1 = tienda['Calificación'].mean()
calificacion_prom_t2 = tienda2['Calificación'].mean()
calificacion_prom_t3 = tienda3['Calificación'].mean()
calificacion_prom_t4 = tienda4['Calificación'].mean()
```

```python
print("La calificación promedio para la tienda 1 es: ", calificacion_prom_t1)
print("La calificación promedio para la tienda 2 es: ", calificacion_prom_t2)
print("La calificación promedio para la tienda 3 es: ", calificacion_prom_t3)
print("La calificación promedio para la tienda 4 es: ", calificacion_prom_t4)
```

# 4. Productos más y menos vendidos

```python
ventas_por_producto_t1 = tienda.groupby('Producto')['Precio'].sum()
ventas_por_producto_t2 = tienda2.groupby('Producto')['Precio'].sum()
ventas_por_producto_t3 = tienda3.groupby('Producto')['Precio'].sum()
ventas_por_producto_t4 = tienda4.groupby('Producto')['Precio'].sum()
```

```python
top_5_ventas_t1 = ventas_por_producto_t1.sort_values(ascending=False).head(5)
top_5_ventas_t2 = ventas_por_producto_t2.sort_values(ascending=False).head(5)
top_5_ventas_t3 = ventas_por_producto_t3.sort_values(ascending=False).head(5)
top_5_ventas_t4 = ventas_por_producto_t4.sort_values(ascending=False).head(5)

print("Los 5 productos más vendidos en la Tienda 1 son:\n", top_5_ventas_t1)
print("\nLos 5 productos más vendidos en la Tienda 2 son:\n", top_5_ventas_t2)
print("\nLos 5 productos más vendidos en la Tienda 3 son:\n", top_5_ventas_t3)
print("\nLos 5 productos más vendidos en la Tienda 4 son:\n", top_5_ventas_t4)
```

##Identificar y mostrar los productos menos vendidos

```python
bottom_5_ventas_t1 = ventas_por_producto_t1.sort_values(ascending=True).head(5)
bottom_5_ventas_t2 = ventas_por_producto_t2.sort_values(ascending=True).head(5)
bottom_5_ventas_t3 = ventas_por_producto_t3.sort_values(ascending=True).head(5)
bottom_5_ventas_t4 = ventas_por_producto_t4.sort_values(ascending=True).head(5)

print("Los 5 productos menos vendidos en la Tienda 1 son:\n", bottom_5_ventas_t1)
print("\nLos 5 productos menos vendidos en la Tienda 2 son:\n", bottom_5_ventas_t2)
print("\nLos 5 productos menos vendidos en la Tienda 3 son:\n", bottom_5_ventas_t3)
print("\nLos 5 productos menos vendidos en la Tienda 4 son:\n", bottom_5_ventas_t4)
```

##Visualizar productos más y menos vendidos

```python
import matplotlib.pyplot as plt


datos_tiendas = {
    "Tienda 1": {
        "top": top_5_ventas_t1,
        "bottom": bottom_5_ventas_t1
    },
    "Tienda 2": {
        "top": top_5_ventas_t2,
        "bottom": bottom_5_ventas_t2
    },
    "Tienda 3": {
        "top": top_5_ventas_t3,
        "bottom": bottom_5_ventas_t3
    },
    "Tienda 4": {
        "top": top_5_ventas_t4,
        "bottom": bottom_5_ventas_t4
    }
}


for store_name, data in datos_tiendas.items():
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    fig.suptitle(f'Ventas de Productos en {store_name}', fontsize=16)


    axes[0].bar(data['top'].index, data['top'].values, color='skyblue')
    axes[0].set_title('Top 5 Productos Más Vendidos')
    axes[0].set_xlabel('Producto')
    axes[0].set_ylabel('Ventas Totales (COP)')
    axes[0].ticklabel_format(style='plain', axis='y')
    axes[0].tick_params(axis='x', rotation=45)


    axes[1].bar(data['bottom'].index, data['bottom'].values, color='red')
    axes[1].set_title('Top 5 Productos Menos Vendidos')
    axes[1].set_xlabel('Producto')
    axes[1].set_ylabel('Ventas Totales (COP)')
    axes[1].ticklabel_format(style='plain', axis='y')
    axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
